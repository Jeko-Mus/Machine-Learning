## 1.- import the Operating System (os) module in python and any other library you need

```python
import os
from PIL import Image
import numpy as np
import pandas as pd
```

## 2.- As you can see each class has its own folder (Do it only for train). 

    - Iterate folder by folder ( os.listdir() )
    - Inside each folder: 
        1.- Read the image
        2.- Reshape it into a flat array (784,)
        3.- Save the data into a pandas dataframe apending the column name as the class
    - Save the data into a CSV

    Note: if it takes to long try doing only 100 images per folder and the teacher for the CSV.

```python
def pathway():
    files = '.'
    for i in os.listdir(files):
        if i == 'trainingSet':
            return i
path = pathway()
```

```python
import os
#path = 'trainingSet'
#files = os.listdir(path)

numbr_images = []
img_names = []
for root, directories, files in os.walk(path, topdown=False):
    for name in files:
        img_names.append(name)
        file_path = os.path.join(root, name)
        j = Image.open(file_path)
        arr = np.array(j, dtype=int)
        arr_flat = arr.flatten()
        numbr_images.append(arr_flat)
```

```python
df = pd.DataFrame(numbr_images)
```

```python
index = pd.Index(img_names)
df = df.set_index(index)
```

### Create Y variable for classes

```python
class_lengths = []
i = 0
while i < 10:
    length = len(os.listdir(f'trainingSet/{i}'))
    class_lengths.append(length)
    i +=1
sum(class_lengths)
```

```python
class_lengths
```

```python
classes = []
for i in os.listdir(path):
    classes.append(i)
classes
```

```python
# get Y values
ys = np.repeat(classes,class_lengths)
ys.shape
```

```python
df['class'] = (ys)
```

## 3.- Load the CSV

```python
df_images = pd.read_csv('num_img_csv', index_col=0)
df_images
```

## 4.- Create a dictionary of models (No preprocessing needed, it has already been done).
    
    Include both, tree models and mult models.

```python
from sklearn.tree          import DecisionTreeRegressor
from sklearn.ensemble      import RandomForestRegressor
from sklearn.ensemble      import ExtraTreesRegressor
from sklearn.ensemble      import AdaBoostRegressor
from sklearn.ensemble      import GradientBoostingRegressor
from xgboost               import XGBRegressor
from lightgbm              import LGBMRegressor
from catboost              import CatBoostRegressor
```

```python
tree_classifiers = {
  "Decision Tree": DecisionTreeRegressor(),
  "Extra Trees":   ExtraTreesRegressor(n_estimators=100),
  "Random Forest": RandomForestRegressor(n_estimators=100),
  "AdaBoost":      AdaBoostRegressor(n_estimators=100),
  "Skl GBM":       GradientBoostingRegressor(n_estimators=100),
  "XGBoost":       XGBRegressor(n_estimators=100),
  "LightGBM":      LGBMRegressor(n_estimators=100),
  #"CatBoost":      CatBoostRegressor(n_estimators=100),
}
```

## 5.- Using either cross validation or stratification find out which is the best model
    - Base your code on the previous two days examples

```python
x = df_images.drop('class', axis = 1)
y = df_images['class']
```

```python
import time
from sklearn import model_selection 
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.metrics import balanced_accuracy_score
```

```python
# since data is huge, will try the models with partial data
partial_x = x[0:1000]
partial_y = y[0:1000]
```

```python
# cross validation

skf = model_selection.StratifiedKFold(n_splits=5, random_state=0, shuffle=True)


results = pd.DataFrame({'Model': [], 'Accuracy': [], 'Bal Acc.': [], 'Time': []})

for model_name, model in tree_classifiers.items():
    start_time = time.time()

# TRAIN AND GET PREDICTIONS USING cross_val_predict() and x,y
    
    pred = cross_val_predict(model, partial_x, partial_y, cv=skf)
    pred = pred.astype('int64')

    total_time = time.time() - start_time

    results = results.append({"Model":    model_name,
                            "Accuracy": metrics.accuracy_score(partial_y, pred)*100,
                            "Bal Acc.": metrics.balanced_accuracy_score(partial_y, pred)*100,
                            "Time":     total_time},
                            ignore_index=True)


results_ord = results.sort_values(by=['Accuracy'], ascending=False, ignore_index=True)
results_ord.index += 1 
results_ord.style.bar(subset=['Accuracy', 'Bal Acc.'], vmin=0, vmax=100, color='#5fba7d')
```

```python
# stratification
x_train, x_val, y_train, y_val = model_selection.train_test_split(partial_x, partial_y, test_size=0.20, random_state=0, stratify = partial_y)

results = pd.DataFrame({'Model': [], 'Accuracy': [], 'Bal Acc.': [], 'Time': []})


for model_name, model in tree_classifiers.items():
    start_time = time.time()

    model.fit(x_train,y_train)

    # GET PREDICTIONS
    pred = model.predict(x_val)
    pred = pred.astype('int64')

    total_time = time.time() - start_time

    results = results.append({"Model":    model_name,
                               "Accuracy": metrics.accuracy_score(y_val, pred)*100,
                               "Bal Acc.": metrics.balanced_accuracy_score(y_val, pred)*100,
                               "Time":     total_time},
                               ignore_index=True)

results_ord = results.sort_values(by=['Accuracy'], ascending=False, ignore_index=True)
results_ord.index += 1 
results_ord.style.bar(subset=['Accuracy', 'Bal Acc.'], vmin=0, vmax=100, color='#5fba7d')
```
